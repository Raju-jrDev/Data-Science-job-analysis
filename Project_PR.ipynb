```python
salary_cleaned = salary_df.drop_duplicates().reset_index(drop=True)
salary_cleaned.duplicated()
```

```python
salary_cleaned = salary_cleaned[~salary_cleaned.index.duplicated(keep='first')]
```

```python
jobtitles = set(salary_cleaned['job_title'].unique())
ET = set(salary_cleaned['employment_type'].unique())
ET = set(salary_cleaned['employment_type'].unique())
EL = set(salary_cleaned['experience_level'].unique())
RR = set(salary_cleaned['remote_ratio'].unique())
CS = set(salary_cleaned['company_size'].unique())
jobtitles,ET,EL,RR,CS
```

```python
employment_mapping = {'FT': 0,'PT': 1,'CT':2,'FL':3}

experience_mapping = {'EN':0, 'EX':1, 'MI':2, 'SE':3}

company_size_mapping = {'S':0, 'M':1, 'L':2}


salary_cleaned['employment_type'] = salary_cleaned['employment_type'].replace(employment_mapping)
salary_cleaned['experience_level'] = salary_cleaned['experience_level'].replace(experience_mapping)
salary_cleaned['company_size'] = salary_cleaned['company_size'].replace(company_size_mapping)
salary_US = salary_cleaned[salary_cleaned['company_location']== 'US']
salary_US = salary_US.drop(['salary','salary_currency'],axis=1)
salary_US

#salary_aggregated = salary_US.groupby(['work_year', 'job_title'], as_index=False).agg({'salary_in_usd':'mean'}).reset_index(inplace=True)
#salary_final = pd.concat([salary_US,salary_aggregated], ignore_index=True)
#salary_final
#salary_final.to_csv('salary_final_PR.csv', index=False)
```

```python
top_roles = salary_US['job_title'].value_counts().nlargest(7).index.tolist()

salary_top7roles = salary_US[salary_US['job_title'].isin(top_roles)]
salary_top7roles.reset_index(inplace=True)
salary_top7roles= salary_top7roles.drop('index',axis=1)
salary_US[salary_US['job_title'] == 'Analytics Engineer']
```

```python
#top_roles = salary_aggregated['job_title'].value_counts().nlargest(5).index.tolist()

#salary_roles_minimized = salary_aggregated[salary_aggregated['job_title'].isin(top_roles)]
#salary_roles_minimized
#salary_cleaned['job_title'].value_counts().nlargest(5)
```

```python
df_pivot = salary_top7roles.pivot_table(index='work_year',columns='job_title',values='salary_in_usd')
df_pivot.plot(kind='line',marker = 'o')

plt.title('Salary trends for top 7 frequent job titles in the dataset')
plt.xlabel('Work Year')
plt.ylabel('Salary in USD')
plt.ylim(50000, 225000)
plt.grid(True)
plt.legend(title='job_title')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
df_pivot = salary_US.pivot_table(index='work_year',columns='remote_ratio',values='salary_in_usd')
df_pivot.plot(kind='line',marker = 'o')

plt.title('Salary trends by Remote Ratio')
plt.xlabel('Work Year')
plt.ylabel('Salary in USD')
plt.ylim(50000, 225000)
plt.grid(True)
plt.legend(title='job_title')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
plt.scatter(salary_US['salary_in_usd'],salary_US['remote_ratio'],color = 'green', marker ='o')
plt.title('Trend of Remote Ratio by Salary')
plt.xlabel('Salary USD')
plt.ylabel('Remote Ratio(%)')
plt.show()
```

```python
# Most highly paind job title
salary_filtered = salary_US[salary_US['salary_in_usd'] >= 350000]
average_salary = salary_filtered.groupby('job_title')['salary_in_usd'].mean().reset_index()
top_roles_salary = average_salary.sort_values(by='salary_in_usd')

plt.figure(figsize=(13,9))
plt.barh(top_roles_salary['job_title'],top_roles_salary['salary_in_usd'],color = 'skyblue')
plt.title('Most Highly Paid Roles')
plt.xlabel('Average Salary(USD)')
plt.ylabel('Job Title')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
companysize_mapping = {
    0: 'Small',
    1: 'Medium',
    2: 'Large'
}
companysize_barplot = salary_top7roles
companysize_barplot = companysize_barplot.sort_values(by='company_size', ascending=False)
companysize_barplot['company_size']=companysize_barplot['company_size'].replace(companysize_mapping)
sns.barplot(x='job_title', y='salary_in_usd', hue='company_size', data=companysize_barplot, palette=('cividis'))
plt.title('Salary by Job Title and Company Size')
plt.xlabel('Job Title')
plt.ylabel('Salary(USD)')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
#To find correlation:
renamed_columns = {
    'work_year':'Work Year',
    'experience_level': 'Experience Level',
    'employment_type' : 'Employment Type',
    'salary_in_usd' : 'Salary',
    'remote_ratio' : 'Remote Ratio',
    'company_size' : 'Company size'}


numeric_df = salary_filtered.select_dtypes(include=['number'])

numeric_df.rename(columns=renamed_columns,inplace=True) 

correlation_matrix = numeric_df.corr()


# Create the heatmap
plt.figure(figsize=(8, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='Greens', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
